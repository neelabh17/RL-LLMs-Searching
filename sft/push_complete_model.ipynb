```python
from peft import PeftModel, PeftConfig
import torch           

from transformers import AutoModelForCausalLM, AutoTokenizer

model_name_or_dir ='anirudhb11/sft-qwen-1.5bi-deg-5-path-5-nodes-300-qwen-14bi-final-all'
peft_config = PeftConfig.from_pretrained(model_name_or_dir)
base_model_name = peft_config.base_model_name_or_path
base_model = AutoModelForCausalLM.from_pretrained(
            base_model_name, torch_dtype=torch.bfloat16
        )
base_model = AutoModelForCausalLM.from_pretrained(base_model_name)
adapter_model = PeftModel.from_pretrained(base_model, model_name_or_dir)
model = adapter_model.merge_and_unload()
tokenizer = AutoTokenizer.from_pretrained('Qwen/Qwen2.5-1.5B-Instruct')
```

```python
import os
SCRATCH_DIR = os.environ['SCRATCH']

output_dir = f'{SCRATCH_DIR}/sft_all'
model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)
```
